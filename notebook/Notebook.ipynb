```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_charts = pd.read_csv('music_charts.csv')
df_incidents = pd.read_csv('incident.csv')
```

```python
print(df_charts.head())
```

```python
print(df_incidents.head())
```

```python
plt.hist(df_charts['valence'], bins=30)
plt.title('Distribution of valence')
plt.xlabel('Valence')
plt.ylabel('Number of songs')
plt.show()
```

```python
plt.plot(df_incidents["date"], df_incidents["incidents"])
plt.xticks(rotation=45)
plt.show()
```
