```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def allsundays(year):
    return pd.date_range(start=str(year), end=str(year+1), 
                         freq='W-FRI').strftime('%Y-%m-%d').tolist()

weeks = allsundays(2023)[:39]  # First 5 Sundays of 2017

def scrapeData(week):
    # Making a GET request
    r = requests.get('https://tophit.com/chart/top/media/hits/ukraine/weekly/' + week)
    print(r)
     
    # Parsing the HTML
    soup = BeautifulSoup(r.content, 'html.parser')
    
    # Locate the element containing the song rows
    # song_rows = soup.find_all('div', class_='Row_name__vEYs2')
    song_rows = soup.find_all('div', class_='Row_rightHolder__Wt9ON')
    
    # Extract and print the song data
    # for song in song_rows:
    #     song_name = song.find('div', class_='flex items-center gap-x-1.5').text.strip()
    #     artist_name = song.find('div', class_='truncate pb-0.5 text-sm leading-4 text-gray-500').text.strip()
    #     isrc = song.find('div', class_='text-gray-500').text.strip()
        
    #     print(f"Song: {song_name}\nArtist: {artist_name}\ISRC: {isrc}\n")

    for song in song_rows:
        # song_name = song.find('div', class_='flex items-center gap-x-1.5').text.strip()
        # artist_name = song.find('div', class_='truncate pb-0.5 text-sm leading-4 text-gray-500').text.strip()
        isrc = song.find('div', class_='text-gray-500').text.strip()
        
        print(f"ISRC: {isrc}\n")

for week in weeks:
    scrapeData(week)
    time.sleep(5)
```

```python
 # Making a GET request
r = requests.get('https://tophit.com/chart/top/media/hits/ukraine/weekly/2023-09-15')
 
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')

# Locate the element containing the song rows
song_rows = soup.find_all('div', class_='Row_name__vEYs2')

# Extract and print the song data
for song in song_rows:
    song_name = song.find('div', class_='flex items-center gap-x-1.5').text.strip()
    artist_name = song.find('div', class_='truncate pb-0.5 text-sm leading-4 text-gray-500').text.strip()
    
    print(f"Song: {song_name}\nArtist: {artist_name}\n")
```
